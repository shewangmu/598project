```python

from icrawler.builtin import BaiduImageCrawler, BingImageCrawler, GoogleImageCrawler
crawl_list = ['dog', 'cat']
for crawler in crawl_list:
    google_crawler = GoogleImageCrawler(
        feeder_threads=2,
        parser_threads=4,
        downloader_threads=16,
        storage={'root_dir': 'dataSet/train/' + crawler + '/'})
    google_crawler.crawl(
        keyword=crawler,
        filters={'date': ((2014, 1, 1), (2014, 6, 30))},
        max_num=200,
        file_idx_offset=0)
    google_crawler = GoogleImageCrawler(
        feeder_threads=2,
        parser_threads=4,
        downloader_threads=16,
        storage={'root_dir': 'dataSet/validation/' + crawler + '/'})
    google_crawler.crawl(
        keyword=crawler,
        filters={'date': ((2016, 1, 1), (2016, 6, 30))},
        max_num=40,
        file_idx_offset=0)
    google_crawler = GoogleImageCrawler(
        feeder_threads=2,
        parser_threads=4,
        downloader_threads=16,
        storage={'root_dir': 'dataSet/test/' + crawler + '/'})
    google_crawler.crawl(
        keyword=crawler,
        filters={'date': ((2017, 1, 1), (2017, 6, 30))},
        max_num=40,
        file_idx_offset=0)
```
